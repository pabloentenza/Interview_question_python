### 1. Get vowel from a sentence

```python
def getVowels(word:str):
    strList = [i for i in word if i in 'aeiou']
    print(f"Vowels: {strList}")

string = "My name is Slim Shady"

getVowels(string)
```

### 2. Capitalizing the First Letter without using title()

```python
def getcapital(word:str):
    print(word)
    wordList = list(word)
    wordList[0] = chr(ord(wordList[0])-32)
    capitalList = "".join([i for i in wordList])
    print(capitalList)

string = "my name is"

getcapital(string)

```

sentence = *Have free hours and love children?*

[('Have', 'free'), ('free', 'hours'), ('hours', 'and'), ('and', 'love'), ('love', 'children')]

```python
sentence = """
Have free hours and love children?
"""

sentenceCleaned = ""

sentenceCleaned = ''.join(letter for letter in sentence if letter not in ["\n","?"])
listaLetter = sentenceCleaned.split()

tuplaList = []
i = 0
while i <= len(listaLetter)-2:
    tupla = (listaLetter[i],listaLetter[i+1])
    tuplaList.append(tupla)
    i+=1

print(tuplaList)
```

###  9. Find Trailing zeros in a factorial of a number

```python
def trailing_zeros_factorial(n: int)->int:

    local_n = n
    factorial = 1
    result_training = 0

    while(local_n > 0):
        factorial = factorial*local_n
        local_n-=1

    print(f"The factorial: {factorial}")
    factorialStr = str(factorial)
    trailing_ct = 0

    for i in range(len(factorialStr)-1):
        if factorialStr[i] != '0' and factorialStr[i+1] == '0':
            trailing_ct += 1

    return trailing_ct


trailing_zeros_factorial(5)
```

###  10. Verify if a word is segmented in a list of words

```python
def verify_dict(s:str, dicti: list)->bool:

    num_items = len(dicti)
    bool_flags = False

    for i in range(num_items-1):
        for j in range(1, num_items):
            if s == "".join([dicti[i],dicti[j]]) or s == "".join([dicti[j],dicti[i]]):
                bool_flags = True

    return bool_flags


s = "dacamp"
dictionary = ["data", "camp", "cam", "lack"]
verify_dict(s, dictionary)
```

### 11. Remove duplicated items from sorted array

```python
def drop_duplicate_number(mList:list)->list:

    mList.sort()   #Ascending order of the list

    mList_rem = [mList[0]]

    for i in range(len(mList)):
      if mList_rem[-1] != mList[i]:
        mList_rem.append(mList[i])


    return mList_rem

lista = [1,1,2,2,2,2,2,2,2,2]
lista  = drop_duplicate_number(lista)
print(lista)
```

###  12. Find missing values in a list where the items are incremented by 1

```python
def find_missing(input_list:list)-> list:

    input_list.sort()
    missing_value = []

    for index in range(len(input_list)-1):

        if  input_list[index+1] != input_list[index]+1:
            i = 1
            diff = input_list[index+1] - input_list[index]
            while( i < diff):
                missing_value.append(input_list[index]+i)
                i += 1

    print(f"Missing values: {missing_value}")

list_1 = [1,2,4,8]

find_missing(list_1)
# 2
```

### 13. Detect pythagorean triplet in an array

```python
def pithagorean_triplet(list_items:list)->bool:
    list_items_c = [i*i for i in list_items]
    list_items_c.sort()

    bool_flag = False
    for i in range(2,len(list_items_c)):
        c2 = list_items_c[i-1] + list_items_c[i-2]
        if list_items_c[i] != c2:
            continue
        else:
            bool_flag = True
            break

    print(bool_flag)

list_items = [10, 4, 6, 12, 5]
pithagorean_triplet(list_items)
```

### 14. Combination of minimum coins that sums a certain value

```python
def coin_combination(sum:int, coins: list)->int:

    arr = [0]*(sum+1)
    arr[0] = 1

    for coin in coins:
        for i in range(1,len(arr)):
            if i >= coin:
               arr[i] += arr[i-coin]


    return arr[-1]



coins = [2,3,5]  # Denomination of coin
SUM = 7          # Total sum of coins

print(coin_combination(SUM, coins))
```

### 15. Given a certain list , find the index (i,j) where the maximum value

```python
def maximun_index(array:list)->tuple:
    max = -1
    index_max = len(array)-1
    index_i = 0
    index_j = 0

    for i in range(len(array)):
        for j in range(index_max,-1,-1):

            if (array[j] - array[i])> max:

                max = array[j] - array[i]
                index_i, index_j = i, j

    return (index_i,index_j)

array =  [20, 70, 40, 50, 12, 38, 98]
print(maximun_index(array))
```

### 16. Verify if a word is an anagram


```python
def letter_count(stg:str)->dict:
  count = {}
  for letter in stg:
    if letter == ' ':
      pass
    elif letter not in count.keys():
      count[letter] = 1
    else:
      count[letter]+=1

  return count

def confirm_anagram(stg1:str, stg2:str)->bool:
  flag = False
  count1 = letter_count(stg1)
  count2 = letter_count(stg2)

  if count1 == count2:
    flag = True

  return flag

word1 = "dracula"
word2 = "alucard"

print(confirm_anagram(word1, word2))
```

### 17. Palindrome checking

```python
def palindrome_cheking(stg:str)->bool:
  flag = False
  invStg = stg[::-1]

  if invStg == stg:
    flag = True

  return flag

word = "aaaabbaaaa"

print(palindrome_cheking(word))
```

### 18. Converting lowecase to uppercase and viceversa

```python
def converting(stg:str):
  converting = ""

  for letter in stg:
    if letter == " ":
      pass
    elif ord(letter) < 90:
      converting+= "".join( chr(ord(letter) + 32))
    else:
      converting+= "".join( chr(ord(letter) - 32))

  print(converting)

name = "Michael Jordan"
converting(name)


```
